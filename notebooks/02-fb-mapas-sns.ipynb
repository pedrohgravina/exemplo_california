```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

from src.config import DADOS_ORIGINAIS, DADOS_LIMPOS
from src.graficos import PALETTE, SCATTER_ALPHA

sns.set_theme(palette="bright")
```

```python
df = pd.read_parquet(DADOS_LIMPOS)

df.head()
```

```python
fig, ax = plt.subplots(figsize=(10, 6))

sns.scatterplot(data=df, x="longitude", y="latitude", ax=ax)

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 6))

sns.scatterplot(data=df, x="longitude", y="latitude", ax=ax, alpha=SCATTER_ALPHA)

plt.show()
```

```python
sns.jointplot(data=df, x="longitude", y="latitude", alpha=SCATTER_ALPHA)
# Vai mostrar um hitograma para cada eixo, melhorando a visualizao da concentrao de registros
```

```python
fig, ax = plt.subplots(figsize=(10, 6))

sns.scatterplot(data=df, x="longitude", y="latitude", ax=ax, alpha=SCATTER_ALPHA, hue="ocean_proximity")
# O hue vai simplesmente mudar as cores de acorodo com a coluna passada pra ele.
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 6))

sns.scatterplot(data=df, x="longitude", y="latitude", ax=ax, hue="median_income_cat", palette=PALETTE)

plt.show()
# Notamos que quanto mais pro interior do estado, mais baratos os imveis.
```

```python
fig, ax = plt.subplots(figsize=(10, 6))

sns.scatterplot(data=df, x="longitude", y="latitude", ax=ax, hue="median_house_value", palette=PALETTE)
# Essa legenda no  a ideal pra esse tipo de grfico, onde a coluna apresenta valores contnuos.
# Ou seja, existem muito mais tons do que os apresentados na legenda.

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 6))

norm_median_house_value = plt.Normalize(df["median_house_value"].min(), df["median_house_value"].max())
# Fiz a normalizao dos dados
sm_median_house_value = plt.cm.ScalarMappable(norm=norm_median_house_value, cmap=PALETTE)
# Fiz o mapeamento scalar dos dados normalizados (associei esses dados a diferentes tons de cores)

sns.scatterplot(data=df, x="longitude", y="latitude", ax=ax, hue="median_house_value", palette=PALETTE)
ax.get_legend().remove()
fig.colorbar(sm_median_house_value, ax=ax)

plt.show()
```
